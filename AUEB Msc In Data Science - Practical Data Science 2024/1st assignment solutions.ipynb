## Neighbourhood Prices in Descending Order

Create a table with the prices per neighbourhood. The prices information should contain:

* median price per neighbourhood

* mean price per neighbourhood

* standard deviation per neighbourhood

* number of listings per neighbourhood.

The contents of the table should be in descending median prices.

```python
# Define aggregation functions

agg_functions = {
    'price': ['median', 'mean', 'std'],
    'id': 'count'
}

# Group by 'Neighborhood' and apply the aggregation functions
result = listings.groupby('neighbourhood').agg(agg_functions)

# Rename the columns for clarity
result = result.rename(columns={'median': 'median_price',
                                'mean': 'mean_price',
                                'std': 'std_dev_price',
                                'count': 'no_of_listings'})
```

```python
# Assuming you have a DataFrame named 'calendar'
q1 = calendar['price'].quantile(0.25)
q3 = calendar['price'].quantile(0.75)

# Calculate the IQR (Interquartile Range)
iqr = q3 - q1

# Define a lower bound and upper bound for outlier detection
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Filter the DataFrame to remove outliers
calendar = calendar[(calendar['price'] >= lower_bound) & (calendar['price'] <= upper_bound)]
```

```python
calendar_with_dates = calendar

# Convert the 'date' column to datetime
calendar_with_dates['date'] = pd.to_datetime(calendar_with_dates['date'])

# Extract the day of the week
calendar_with_dates['day_of_week'] = calendar_with_dates['date'].dt.day_name()

# Remove the dollar sign and convert 'price' to numeric
calendar_with_dates['price'] = pd.to_numeric(calendar_with_dates['price'].str.replace('$', '').str.replace(',', ''), errors='coerce')



```

```python
# Assuming you have a DataFrame named 'calendar_with_dates'

# First, convert the "price" column to string (if not already)
calendar_with_dates['price'] = calendar_with_dates['price'].astype(str)

# Then, remove the dollar sign and commas
calendar_with_dates['price'] = calendar_with_dates['price'].str.replace('$', '').str.replace(',', '')

# Now, convert the cleaned "price" column to numeric
calendar_with_dates['price'] = pd.to_numeric(calendar_with_dates['price'], errors='coerce')
```

```python

# Group by 'listing_id' and 'day_of_week', calculate the average price
result = calendar_with_dates.groupby(['listing_id', 'day_of_week'])['price'].mean().reset_index()

sns.catplot(data=result, x="price", y="day_of_week")
```
